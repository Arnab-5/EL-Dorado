These apps have come to play a huge role in the way we live our lives today. Our Objective is to find the Most Popular Category, find the App with largest number of installs , the App with largest size etc.

```python
import numpy as np


df['Size'] = (
    df['Size']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('+', '', regex=False)
    .str.replace('M', '000', regex=False)
    .str.replace('k', '', regex=False)
    .replace('Varies with device', np.nan)
    .astype(float)
)
```

```python
df.head()
```

```python
df['Installs']=(df['Installs'].str.replace(',', '')
    .str.replace('+', '')
    .str.replace('Free','0').astype(int))
```

```python
df.head()
```

```python
df.groupby('Category').agg(
    total_apps=('Category', 'count'),
    avg_rating=('Rating', 'mean')
).sort_values('total_apps', ascending=False)
```
